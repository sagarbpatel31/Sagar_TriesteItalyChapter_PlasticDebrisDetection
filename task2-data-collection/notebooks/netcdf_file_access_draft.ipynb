# Convert netcdf to DataFrame and save as csv

Code to read raw data in netcdf format from Cozar & Arias et al. (2024) *'Proof of concept for a new sensor to monitor marine litter from space'*, published in Nature Communications. <br> Data retrieved from https://zenodo.org/records/11045944 <br>
This data will be used to create masks for future algorithms

This dataset contains the 14,374 Litter Windrows detections found using multispectral Copernicus Sentinel-2/MSI L1c data in the Mediterranean Sea for the period June 2015–September 2021. The dataset includes exclusively the filaments deemed as valid by the research team and used in the main research of the associated research paper. The following information is provided:

- Geocentric lat/lon coordinates (WGS84) of the 'centroid' position for the filament
- (X, Y) coordinates of the filament's centroid related to the multispectral Sentinel-2 image array, when considering all spectral bands resampled at 10m spatial resolution.
- Number of 10x10m pixels composing the filaments where floating matter has been identified.
- (X, Y) coordinates for all the detected pixels within the filament and related to the associated Sentinel-2 L1c image at 10m resolution.
- Original full spectrum (13 bands) for each of the pixels within the filament.
- Limits of the bounding box (in image (x, y) coordinates at 10m resolution) contain the entire filament.
- Full name of the Sentinel-2 L1c product containing the filament.
- Central decimal time (UTC) of the associated Sentinel-2 acquisition.
